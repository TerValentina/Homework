```python
import matplotlib.pyplot as plt
import numpy as np
a = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
fig = plt.figure(figsize = (10, 10))
fig.patch.set_facecolor('xkcd:white')
#X = np.arange(0, 21, dtype = int)
H, L, C = a[:, 1], a[:, 2], a[:, 3]
plt.plot(np.argmax(a[:, 1]) + 1900, a[np.argmax(a[:, 1]), 1] , '-x')
plt.plot(np.argmax(a[:, 2]) + 1900, a[np.argmax(a[:, 2]), 2] , '-x')
plt.plot(np.argmax(a[:, 3]) + 1900, a[np.argmax(a[:, 3]), 3] , '-x')
plt.plot(a[:, 0], H, label='Hares', c = 'b')
plt.plot(a[:, 0], L, label='Lynxes', c='g')
plt.plot(a[:, 0], C, label='Carrots', c='r')
plt.xlabel('Years', fontsize=16)
plt.ylabel('Population')
plt.xticks(fontsize=14)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
fig = plt.figure(figsize = (10,10))
fig.patch.set_facecolor('xkcd:white')
d1 = np.diff(a[:, 1])

newa = np.delete(a[:, 2], 20)
newb = np.delete(a[:, 0], 20)

plt.plot(newb, d1)
plt.plot(newb, newa)

cor = np.corrcoef(d1, newa)
print('coef = ', cor[0, 1])

plt.legend(['Hare','Lynx'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```
